```python
import requests 
from itertools import count
from bs4 import BeautifulSoup 
```

```python
## Finding Quotes by Author 

## Defining url, getting page with requests and defining the "soup"
url = "https://quotes.toscrape.com/"

page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

## Getting Albert Einstein quotes, for the first page 
desired_author = "Albert Einstein"
all_quotes = soup.find_all('div', class_="quote")

desired_author_quotes = []
for quote in all_quotes: 
    quote_author_html = quote.find(class_="author")
    if(quote_author_html):
        quote_author = quote_author_html.text 
        if(quote_author == desired_author):
            quote_text = quote.find(class_="text").text
            desired_author_quotes.append(quote_text)
            
for d_quote in desired_author_quotes:
    print(d_quote, "- {author}".format(author=desired_author))
```

```python
## Let's say we wanted to get all quotes for a specific tag

## First, we got find all possible tags. So, let's go through all pages and collect all unique tags 
tags = []
for page in count(1):
    current_url = "https://quotes.toscrape.com/page/{current_page}/".format(current_page=page)
    print(current_url)
    current_page_response = requests.get(current_url)
    current_soup = BeautifulSoup(current_page_response.content, "html.parser")
    ## Validating if it current page is still a valid page
    validation_quote = current_soup.find('div', class_="quote")
    if(validation_quote == None):
        print("No more quotes to find ! \n")
        break
    
    result_tags = current_soup.find_all('a', class_="tag")
    for tag_html in result_tags:
        ## Validating if tag_html isn't a one of the "Top ten tags" element
        parent_class = tag_html.parent.attrs["class"][0]
        if(parent_class == "tag-item'"): 
            continue 
        tag = tag_html.text
        ## Validating if tag is already in tags list
        if(tag in tags):
            continue
        ## Passed all validations, apppending tag to list
        tags.append(tag)
print("Printing all possible tags \n")
for tag in tags:
    print(tag)
```

```python
## Now, let's get all quotes for a given tag: "inspirational"
## Although we could go through all pages again and extract the quotes that apply to this tag, there is a better approach
## This site provides specific quote filtering base on url paths, so, for example, in case of inspirational quotes:
## The path would be: https://quotes.toscrape.com/tag/inspirational/
## That explained, now let's iterate for all pages possible for inspirational quotes:
desired_tag = "inspirational"
desired_tag_quotes = []
for page in count(1):
    current_url = "https://quotes.toscrape.com/tag/{tag}/page/{current_page}/".format(tag=desired_tag, current_page=page)
    print(current_url)
    current_page_response = requests.get(current_url)
    current_soup = BeautifulSoup(current_page_response.content, "html.parser")
    ## Validating if it current page is still a valid page
    all_quotes = current_soup.find_all('div', class_="quote")
    if(len(all_quotes) == 0):
        print("No more quotes to find ! \n")
        break
    for quote in all_quotes:
        quote_text = quote.find(class_="text").text
        desired_tag_quotes.append(quote_text)

print("Printing quotes for tag {filtering_tag}: \n".format(filtering_tag=desired_tag))
for d_quote in desired_tag_quotes:
    print(d_quote, "\n")
```
